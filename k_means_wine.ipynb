```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./wine.csv')
display(df.head())
np_array = df.to_numpy()
# print(np_array)
```

```python
# display(df.head())
df_alcohol_missing = df[df['alcohol'].isnull()]
df_alcohol_clean = df[df['alcohol'].notnull()]
display(df_alcohol_clean.describe())
```

```python
features = ['fixed_acidity', 'residual_sugar', 'density']  # Add more if needed
X = df_alcohol_clean[features]
y = df_alcohol_clean['alcohol']
```

```python
model = RandomForestRegressor()
model.fit(X, y)

# 4. Predict and fill missing
predicted_alcohol = model.predict(df_alcohol_missing[features])
df.loc[df['alcohol'].isnull(), 'alcohol'] = predicted_alcohol
df.describe()
```

```python
df_sugar_missing = df[df['residual_sugar'].isnull()]
df_sugar_clean = df[df['residual_sugar'].notnull()]
display(df_sugar_clean.describe())
```

```python
sugar_model = RandomForestRegressor()
sugar_model.fit(X, y)

# 4. Predict and fill missing
predicted_sugar = sugar_model.predict(df_sugar_missing[features])
df.loc[df['residual_sugar'].isnull(), 'residual_sugar'] = predicted_sugar
df.describe()
```

```python
features = ['alcohol', 'fixed_acidity', 'residual_sugar', 'density']
X = df[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
pd.crosstab(df['cluster'], df['quality'])
```
